```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("hiring.csv")
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.rename(columns={"test_score(out of 10)":"test_score"},inplace=True)
df
```

```python
df.test_score = df.test_score.fillna(df.test_score.mean())
df
```

```python
df.rename(columns={"interview_score(out of 10)":"interview_score"},inplace=True)
df
```

```python
df.to_csv("hiring_new.csv", index=False)
```

```python
df2 = pd.read_csv("hiring_new.csv")
df2
```

```python
df2.rename(columns = {"salary($)":("salary")}, inplace = True)
df2
```

```python
df2.to_csv("hiring_new.csv", index=False)
```

```python
df3 = pd.read_csv("hiring_new.csv")
df3
```

```python
df3['experience'] = df3['experience'].apply(lambda x: w2n.word_to_num(x))

df3
```

```python
reg = linear_model.LinearRegression()
reg.fit(df3[['experience','test_score','interview_score']],df3.salary)
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12, 10, 10]])
```
